## Preparing Main Result Table

Download the spreadsheet as is as excel file, put it inside ```data/```<br>
Run the below cells

````python
import pandas as pd

path = 'data/'
````

````python
sheet_dict = {'boom_nlu':'Boomer NLU',
              'boom_nlg':'Boomer NLG',
              'ft_nlu':'FT NLU Result',
              'ft_nlg':'FT NLG Result',
              'cls_nlg':'Classic NLG Result',
              'zs_nlu':'ZS NLU Result',
              'zs_nlg':'ZS NLG Result',}
````

````python
sheet_name = sheet_dict['boom_nlu']
metric = 'macro-F1'

df = pd.read_excel(path+'[NusaMenulis] Experiment Log.xlsx',
                      sheet_name=sheet_name)

df_grouped = df.groupby(['Dataset', 'Model', 'Task'])[metric].mean().reset_index()

df_nusa_alinea = df_grouped[df_grouped['Dataset'] == 'nusa_alinea'].drop('Dataset', axis=1)
df_nusa_kalimat = df_grouped[df_grouped['Dataset'] == 'nusa_kalimat'].drop('Dataset', axis=1)

pivot_df_nusa_alinea = df_nusa_alinea.pivot(index='Model', columns='Task', values=metric)
pivot_df_nusa_alinea = pivot_df_nusa_alinea.rename_axis(None)
pivot_df_nusa_alinea.columns.name = 'Model'

pivot_df_nusa_kalimat = df_nusa_kalimat.pivot(index='Model', columns='Task', values=metric)
pivot_df_nusa_kalimat = pivot_df_nusa_kalimat.rename_axis(None)
pivot_df_nusa_kalimat.columns.name = 'Model'

pd.concat([pivot_df_nusa_alinea,pivot_df_nusa_kalimat],axis=1,keys=['nusa_alinea','nusa_kalimat']).sort_index(axis=1)
````

````python
sheet_name = sheet_dict['zs_nlu']
metric = 'macro_f1'

df = pd.read_excel(path+'[NusaMenulis] Experiment Log.xlsx',header=[0,1],
                      sheet_name=sheet_name)

df_performance = df['mean'][metric]
df=df[['dataset', 'task', 'model', 'lang']]
df.columns = df.columns.droplevel(1)
df = df.merge(df_performance, left_index=True, right_index=True)

df_grouped = df.groupby(['dataset', 'model', 'task'])[metric].mean().reset_index()

df_nusa_alinea = df_grouped[df_grouped['dataset'] == 'nusa_alinea'].drop('dataset', axis=1)
df_nusa_kalimat = df_grouped[df_grouped['dataset'] == 'nusa_kalimat'].drop('dataset', axis=1)

pivot_df_nusa_alinea = df_nusa_alinea.pivot(index='model', columns='task', values=metric)
pivot_df_nusa_alinea = pivot_df_nusa_alinea.rename_axis(None)
pivot_df_nusa_alinea.columns.name = 'model'

pivot_df_nusa_kalimat = df_nusa_kalimat.pivot(index='model', columns='task', values=metric)
pivot_df_nusa_kalimat = pivot_df_nusa_kalimat.rename_axis(None)
pivot_df_nusa_kalimat.columns.name = 'model'

pd.concat([pivot_df_nusa_alinea,pivot_df_nusa_kalimat],axis=1,keys=['nusa_alinea','nusa_kalimat']).sort_index(axis=1)
````

````python
sheet_name = sheet_dict['zs_nlg']

df = pd.read_excel(path+'[NusaMenulis] Experiment Log.xlsx',header=[0,1],
                      sheet_name=sheet_name)
df_performance = df['mean']
df=df[['dataset', 'task', 'model', 'src_lang', 'tgt_lang']]
df.columns = df.columns.droplevel(1)
df = df.merge(df_performance, left_index=True, right_index=True)

df_grouped = df.groupby(['dataset', 'model']).mean().reset_index()
df_nusa_kalimat = df_grouped[df_grouped['dataset'] == 'nusa_kalimat'].drop('dataset', axis=1)
df_nusa_kalimat.head()
````
